## __Copules et modélisation de la dépendance entre variable aléatoires__

*__Pour $d \in \mathbb{N}, d ≥ 2$, on appelle copule de dimension $d$ une fonction $C : [0, 1]^d \rightarrow [0, 1]$ telle qu’il existe des variables aléatoires $U_1, \dots ,U_d$ uniformes sur $[0, 1]$ avec lesquelles $$\forall u \in [0, 1]^d, C(u_1, \dots , u_d) = \mathbb{P}(U_1 ≤ u_1, \dots ,U_d ≤ u_d)$$ Ainsi, une copule est la fonction de répartition d’un vecteur aléatoire dont les marginales suivent la loi uniforme sur $[0, 1]$.__*

*__On définit $C^+(u) = \min_{1≤i≤d} \ u_i$ et $C^-(u) = max(\sum_{i=1}^{d} u_i − (d − 1), 0)$. Montrer que $$\forall u \in [0, 1]^d, \ C^−(u) ≤ C(u) ≤ C^+(u)$$ A quelle condition a-t-on $C = C^+$? (Pour la minoration par $C^−$, on pourra commencer par le cas $d$ = 2 et faire une récurrence.)__*


Soit $\space u \in [0,1]^d$, 

On a pour tout $ 1 \leq i \leq n , \space (U_1 \leq u_1,\ldots, U_d \leq u_d) \subset (U_i \leq u_i)$ donc, $C(u) \leq \mathbb{P}(U_i \leq u_i)=u_i$ . Ceci étant vrai pour tout $i$ on en déduit que $C(u) \leq C^+(u) $

Pour la deuxième inégalité, nous allons effectuer une récurrence sur $d\geq 2$ :

Commençons par le cas $d=2$, on a $C^-(u)=\max(u_1+u_2-1,0)$, clairement si $C^-(u)=0$ on a bien $C^-(u) \leq C(u)$, dans l'autre cas, c'est-à-dire si $C^-(u)=u_1+u_2 -1$, alors comme $C(u)=u_1+u_2-\mathbb{P}((U_1 \leq u_1) \cup (U_2 \leq u_2) ) $ et que $\mathbb{P}((U_1 \leq u_1) \cup (U_2 \leq u_2) ) \leq 1$ on obtient $C(u) \geq C^-(u)$

Si maintenant l'inégalité est vraie pour toute copule de dimension $d\geq 2$, on se donne $C$ une copule de dimension $d+1$. Soit $u \in [0,1]^{d+1}$, si $C^-(u)=0$ le résultat est immédiat. Sinon, on a 
$$
\begin{aligned}
C(u) &= \mathbb{P}\Big((U_1 \leq u_1, \ldots, U_d \leq u_d) \cap (U_{d+1} \leq u_{d+1})\Big) \\
&\geq \mathbb{P}(U_1 \leq u_1, \ldots, U_d \leq u_d) + \mathbb{P}(U_{d+1} \leq u_{d+1}) - 1 \\
&\geq \sum_{i=1}^d u_i -(d-1) + u_{d+1} - 1 \quad \text{(par hypothèse de récurrence)} \\
&\geq C^-(u)
\end{aligned}
$$
On en déduit alors que:
$$\boxed{\forall d \geq 2, \text { pour toute copule de dimension d}, \forall u \in [0,1]^d, C^-(u) \leq C(u)\leq C^+(u)}$$

#### __Question T4.__

*__Vérifier que lorsque $d = 2$, $(U_1,U_2)$ a pour copule $C^−$ si et seulement si $\mathbb{P}(U_1 = 1 − U_2) = 1$. En déduire que pour $d ≥ 3$, $C^−$ n’est pas une copule.__*

On se place ici dans le cas $d=2$. Je me donne $(U_1,U_2)$ un vecteur aléatoire de marginales suivant chacune une loi uniforme sur $[0,1]$. $\newline$


Supposons  que $\mathbb{P}(U_1=1-U_2)=1$,

$$\begin{aligned}
\mathbb{P}(U_1\leq u_1,U_2\leq u_2)&=\mathbb{P}(U_1\leq u_1,U_2\leq u_2,U_1=1-U_2)+\mathbb{P}(U_1\leq u_1,U_2\leq u_2,U_1\not=1-U_2))\\
                                   &=\mathbb{P}(U_1\leq u_1,U_2\leq u_2|U_1=1-U_2)\mathbb{P}(U_1=1-U_2)\\
                                   &=\mathbb{P}(U_1\leq u_1,1-u_2\leq U_1)\\
                                   &=\begin{cases} 0 & \text{si } u_1\leq 1-u_2\\ u_1+u_2-1 & \text{sinon }  \end{cases}\\
                                   &=\max(u_1+u_2-1,0)\\
                                   &\boxed{=C^-(u_1,u_2)}\\
\end{aligned}                              
$$

Réciproquement, considérons maintenant que $(U_1,U_2)$ a pour copule $C^-$, je vais faire un raisonnement un peu géométrique. 
Au vu de la copule, on a : la loi du couple est nulle sur le triangle $\{(x,y) \in [0,1] \mid x+y < 1 \}$. Pour le triangle $\{(x,y) \in [0,1] \mid x+y > 1 \}$, calculons:

$$
\begin{aligned}
\mathbb{P}(U_1 > u_1,U_2 > u_2) &= \mathbb{P}((U_1 > u_1) \cap (U_2 > u_2)) \\
                                &= 1 - \mathbb{P}(((U_1 > u_1) \cap (U_2 > u_2))^c) \\
                                &= 1 - \mathbb{P}((U_1 \leq u_1) \cup (U_2 \leq u_2)) \\
                                &= 1 - \mathbb{P}((U_1 \leq u_1) + (U_2 \leq u_2) - ((U_1 \leq u_1) \cap (U_2 \leq u_2))) \\
                                &= 1 - u_1 -u_2 + (u_1 + u_2 - 1) \\
                                &= 0
\end{aligned}
$$

Ainsi, comme $\mathbb{P}(U_1 + U_2 < 1) = \mathbb{P}(U_1 + U_2 > 1) = 0$, on conclut que:

$$\fbox{$\mathbb{P}(U_1 = 1 - U_2) = 1$}$$

Donc, on termine la preuve de l'équivalence:

$$\fbox{Pour $d=2$, $(U_1, U_2)$ a pour copule $C^-$ si, et seulement si $\mathbb{P}(U_1 = 1 - U_2) = 1$}$$
De plus, pour $d \geq 3$, supposons que $C^-$ soit la copule de $(U_1, \dots, U_d)$. D'après la question $T1$, on sait que les copules $C_{U_1, U_2}$, $C_{U_1, U_3}$ et $C_{U_2, U_3}$ de $(U_1, U_2)$, $(U_1, U_3)$ et $(U_2, U_3)$, respectivement, sont telles que:

$$
\begin{align}
    C_{U_1, U_2} = C^-(u_1, u_2, 1, \dots, 1) = \max(u_1 + u_2 -1, 0) \\
    C_{U_1, U_3} = C^-(u_1, 1, u_3, \dots, 1) = \max(u_1 + u_3 -1, 0) \\
    C_{U_2, U_3} = C^-(1, u_2, u_3, \dots, 1) = \max(u_2 + u_3 -1, 0) \\
\end{align}
$$

Ainsi, nous nous sommes ramenés au cas de dimension 2, pour lequel nous avons démontré que:

$$\mathbb{P}(U_1 = 1 - U_2) = 1, \ \mathbb{P}(U_1 = 1 - U_3) = 1, \ \mathbb{P}(U_2 = 1 - U_3) = 1$$

Ceci implique que $\mathbb{P}(U_2 = U_3) = 1 \Rightarrow U_2 = U_3$ presque-sûrement et $U_2 = 1 - U_3$ presque-sûrement. En comparant ces résultats, nous conclusons que $U_2 = 0.5$ presque sûrement. Or, cela est une absurde car $U_2 \sim \mathcal{U}([0, 1])$, par construction de la copule. Donc:

$$\fbox{Pour $d \geq 3$, $C^-$ n'est pas une copule}$$


#### __Question T5.__

*__Montrer que $C$ est une fonction continue.__*


Montrons que C est lipschitzienne : Soit $u,v \in [0,1]^d$:

$$\begin{aligned}
|C(u)-C(v)|& =|\mathbb{P}(U_1 \leq u_1,\ldots,U_d \leq u_d)-\mathbb{P}(U_1 \leq v_1,\ldots,U_d \leq v_d)|\\
           &=\mathbb{P}((\min(u_1,v_1)\leq U_1\leq \max(u_1,v_1),\ldots,\min(u_d,v_d)\leq U_d\leq \max(u_d,v_d))) \\
           &\leq \sum_{i=1}^d \mathbb{P}(\min(u_i,v_i)\leq U_i\leq \max(u_i,v_i))\\
           &\leq \sum_{i=1}^d |u_i-v_i|\\
           &\leq \|u-v\|_1\\
\end{aligned}$$

$$\fbox{On en conclut que C est lipschitzienne donc continue.}$$

#### __Question T6.__

*__Soient $V_1, \dots , V_d$ des variables aléatoires uniformes sur $[0, 1]$ de copule $C^′$ et $B$ une variable aléatoire de Bernoulli de paramètre $p$ indépendante de $U$ et $V$ . Montrer que $BU + (1 − B)V$ est un vecteur aléatoire dont les marginales suivent la loi uniforme sur $[0, 1]$ de copule $pC + (1 − p)C^′$. En déduire que l’ensemble des copules est convexe.__*

Commençons par montrer que les marginales de $BU+(1-B)V$ suivent la loi uniforme sur $[0,1]$:

$BU+(1-B)V $ est à valeur dans $[0,1]$. Soit alors $ 1 \leq i \leq d$ et $x \in [0,1]$,

$$\begin{align*}
F_{BU_i+(1-B)V_i}(x) &= \mathbb{P}(BU_i+(1-B)V_i \leq x) \\
                 &= \mathbb{P}(V_i \leq x, B=0) + \mathbb{P}(U_i\leq x, B=1) \space \text{ par la formule des probabilités totales}\\
                 &= \mathbb{P}(V_i\leq x) \mathbb{P}(B=0) + \mathbb{P}(U_i\leq x) \mathbb{P}(B=1) \space \text{ par indépendance de B avec U et V} \\
                 &= (1-p)x + px\\
                 &= x
                  \end{align*}$$

On en déduit que pour tout $ x \in \mathbb{R}$, $F_{BU_i+(1-B)V_i}(x)=x \mathbb{1}_{x\in [0,1]}$. Ainsi, comme la fonction de répartition caractérise entièrement la loi, on en déduit que $BU_i+(1-B)V_i$ suit une loi uniforme sur $[0,1]$. $\newline$
Soit $(u_1,\ldots,u_d) \in [0,1]^d$, 
$$\begin{align*}
\mathbb{P}(BU_1+(1-B)V_1 \leq u_1, \ldots,BU_d+(1-B)V_d) &=\mathbb{P}(V_1 \leq u_1,\ldots, V_d \leq u_d, B=0) +\mathbb{P}(U_1 \leq u_1,\ldots, U_d \leq u_d, B=1) \\
                                                         &= (1-p)C'(u)+pC(u) \text{  par indépendance de B avec U et V}
                                                         \end{align*}$$

$$
\fbox{\text{On peut donc conclure que } $pC + (1-p)C'$ \text{ est une copule associée au vecteur aléatoire } $BU + (1-B)V$.}
$$

Ainsi, si C et $C'$ sont des copules, $\begin{equation} \forall p \in [0,1] \space pC+(1-p)C' \end{equation}$ est une copule.

$$\fbox{On en conclut que l'ensemble des copules est convexe.}$$

### __Le théorème de Sklar__

*__On considère $d$ variables aléatoires réelles $X_1, \dots , X_d$. Pour $i \in {1, \dots , d}$, on note $F_i(x) = \mathbb{P}(X_i ≤ x)$ la fonction de répartition de $X_i$, que l’on suppose continue dans tout ce paragraphe.__*

*__L’objectif de cette partie est de montrer qu’il existe une unique copule $C_X : [0, 1]^d \rightarrow [0, 1]$ telle que $$\forall x \in \mathbb{R}^d, \mathbb{P}(X_1 ≤ x_1, . . . , X_d ≤ x_d) = C_X(F_1(x_1), \dots , F_d(x_d)) \ \ (7.1)$$ La fonction $C_X$ est appelée la copule du vecteur $X = (X_1, \dots , X_d)$.__*

#### __Question T7.__

*__On commence par supposer les fonctions $F_i$ strictement croissantes, et on pose $U_i = F_i(X_i)$. Montrer que $U_1, \dots ,U_d$ suivent la loi uniforme sur $[0, 1]$. Montrer que la copule associée à $U_1, \dots ,U_d$ est l’unique fonction continue $C_X : [0, 1]^d \rightarrow [0, 1]$ satisfaisant (7.1).__*



Les $F_i$ sont continues et strictement croissantes, elles réalisent donc une bijection de $\mathbb{R}$ dans $[0,1]$ et les $F_i^{-1}$ sont strictement croissantes.

Soit $1 \leq i \leq d$, $U_i$ est à valeur dans $[0,1]$, soit $x \in [0,1]$,

$$
\begin{aligned}
F_{U_i}(x) &= \mathbb{P}(U_i \leq x) \\
           &= \mathbb{P}(F_i(X_i) \leq x) \\
           &= \mathbb{P}(X_i \leq F_i^{-1}(x)) \\
           &= F_i(F_i^{-1}(x)) \\
           &= x
\end{aligned}$$

$$\fbox{Ainsi, $U_i$ suit bien une loi uniforme sur [0,1].}$$


Soit $x\in \mathbb{R}^d$,
$$ \begin {aligned}
C_X(F_1(x_1),\ldots,F_d(x_d))&=\mathbb{P}(U_1 \leq F_1(x_1),\ldots,U_d\leq F_d(x_d))\\
                            &=\mathbb{P}(F_1(X_1)\leq F_1(x_1),\ldots,F_d(X_d)\leq x_d)\\
                            &=\mathbb{P}(X_1\leq x_1, \ldots,X_d\leq x_d) \text{  en composant par les $F_i^{-1}$ qui sont strictement croissantes}\\
                            \end{aligned}$$
$$\fbox{Ainsi, $C_X$ est une fonction continue qui vérifie l'équation $(7.1)$}$$

Soit $C : [0,1]^d \mapsto [0,1]$ continue et vérifiant $(7.1)$,
Soit $u\in [0,1]^d$,
$$ \begin{aligned}
C(u) &= C(F_1(F_1^{-1}(u_1)),\ldots,F_d(F_d^{-1}(u_d))\\
     &= \mathbb{P}(X_1\leq F_1^{-1}(u_1),\ldots,X_d \leq F_d^{-1}(u_d))\\
     &=C_X(F_1(F_1^{-1}(u_1)),\ldots,F_d(F_d^{-1}(u_d))\\
     &=C_X(u)
     \end{aligned}$$
$$\fbox{D'où $C=C_X$ et on conclut l'unicité.}$$

#### __Question T8.__

*__On ne suppose plus les fonctions de répartitions strictement croissantes et on pose pour $u \in ]0, 1[$, $$F^{−1}_i(u) = \inf\{x \in \mathbb{R}, F_i(x) > u\}.$$ Montrer que $F_i(F^{−1}_i(u)) = u$, puis que $F_i(x) ≤ u \Leftrightarrow x ≤ F^{−1}_i(u)$. Ensuite, par définition de $F_i^{-1}(u)$ on a pour tout $x <F_i^{-1}(u)$ : $ F_i(x) \leq u$. Mais alors, en faisant tendre x vers $F_i^{-1}(u)$ par valeur inférieure et en utiisant la continuité de $F_i$, on obtient $F_i(F_i^{-1}(u))\leq u$. $\newline$ D'où: $ \boxed{F_i(F_i^{-1}(u))= u }$.

Donc $\boxed{\forall x \in \mathbb{R}, \space \forall u\in ]0,1],\space F_i(x) \leq u \ \Leftrightarrow\ x \leq F_i^{-1}(u)}$

           &= \mathbb{P}(F_i(X_i) \leq x) \\
           &= \mathbb{P}(X_i \leq F_i^{-1}(x)) \text{  par l'équivalence précédente}\\

                                          \end{aligned}
$$

On conclut l'unicité avec le même calcul que la question 7.

$$\fbox{Ainsi, la copule associée à $(U_1,\ldots,U_d)$ est l’unique fonction continue $C_X : [0, 1]^d \longrightarrow [0, 1]$ satisfaisant (7.1)}.$$

Soit $u=(u_1,\ldots,u_d)\in[0,1]^d$, comme les $F_{X_i}$ sont continues, que $\lim_{-\infty} F_{X_i} =0$ et $\lim_{+\infty} F_{X_i} =1$ le théorème des valeurs intermédiaires assure l'existence de $x_i$ tel que $F_{X_i}(x_i)=u_i$

$$\begin{aligned}
C_{X}(u) &=\mathbb{P}(X_1\leq x_1,\ldots,X_d\leq x_d)\\
            &=\mathbb{P}(f_1(X_1)\leq f_1(x_1),\ldots, f_d(X_d)\leq f_d(x_d)) \text{  par stricte croissance des $f_i$}\\
            &=C_{f(X)}(\mathbb{P}(f_1(X_1)\leq f_1(x_1)),\ldots,\mathbb{P}(f_d(X_d)\leq f_d({x_d})))\\
            &=C_{f(X)}(\mathbb{P}(X_1\leq x_1),\ldots,\mathbb{P}(X_d\leq x_d))\\
            &=C_{f(X)}(u_1,\ldots,u_d)\\
            &=C_{f(X)}(u)
\end{aligned}$$

$$\fbox{Ainsi, les vecteurs $(X_1,\ldots,X_d)$ et $(f_1(X_1),\ldots,F_d(X_d))$ ont la même copule.}$$

### __Copules gaussiennes__

__On appelle copule gaussienne de dimension $d$ la copule d’un vecteur gaussien de dimension $d$. Sans perte de généralité, on supposera que chaque coordonnée de ce vecteur a une variance strictement positive.__

#### __Question T11.__

__*Montrer que pour toute copule Gaussienne, il existe un vecteur gaussien $X$ de dimension $d$, centré et de matrice de covariance $\Sigma$ telle que $\Sigma_{ii} = 1$ pour $1 ≤ i ≤ d$ qui a cette copule.*__

Par définition, une copule gaussienne de dimension $d$ est la copule d'un vecteur gaussien de dimension $d$. Nous allons dériver une méthode pour obtenir un vecteur gaussien centré dont la matrice de covariance a des éléments diagonaux égaux à 1 et qui possède la même copule que le vecteur initial.

Considérons une copule gaussienne arbitraire $C$ et notons $X$ son vecteur gaussien associé. Ce vecteur suit une loi normale $\mathcal{N}_d(m, \Sigma)$, où $m$ est le vecteur moyenne et $\Sigma$ la matrice de covariance, avec chaque variance marginale supposée strictement positive.

Comme $\Sigma$ est symétrique et semi-définie positive par construction, elle est diagonalisable. On peut donc obtenir une racine carrée symétrique $\Sigma^{1/2}$ (dont la construction sera détaillée dans la partie S1). Appliquons ensuite à $X$ la transformation affine croissante composante par composante :

$$f_i(x) = \frac{x - m_i}{\sigma_i}$$

où $\sigma_i = \sqrt{\Sigma_{ii}}$ est l'écart-type de la $i$-ème composante. D'après la proposition 6.13 du polycopié et notre démonstration dans la question T10, le nouveau vecteur $X'$ ainsi obtenu vérifie les propriétés qu'on cherche: 

1. Il est centré : $m' = 0$.
2. Sa matrice de covariance $\Sigma'$ a des éléments diagonaux unitaires.
3. Cette transformation préserve la copule associée, alors $X'$ a $C$ comme copule.

Donc, cette méthode nous permet de conclure que: 

$$\fbox{Toute copule Gaussienne peut être obtenue à partir d’un vecteur gaussien centré dont la matrice de covariance a des diagonales égales à 1}.$$



#### __Simulation S1.__

*__Donner une méthode de simulation pour $X \sim \mathcal{N}_d(0, \Sigma)$ avec $\Sigma_{ii} = 1$ pour $1 ≤ i ≤ d$. (Rappeler comment simuler la loi gaussienne centrée réduite $\mathcal{N}(0, 1)$ à partir d’une loi uniforme sur $[0, 1]$, puis expliquer comment construire $X$ à partir de $d$ variables aléatoires gaussienne centrée réduite indépendantes.)__* 

*__On note $\Phi(x)$ la fonction de répartition de la loi gaussienne centrée et réduite. Montrer alors que $(\Phi(X_1), \dots , \Phi(X_d))$ est un vecteur dont la fonction de répartition est la copule gaussienne de matrice de covariance Σ.__*



Une méthode pour simuler $X \sim \mathcal{N}_d(0, \Sigma)$ consiste, tout d'abord, à simuler $d$ variables aléatoires indépendantes de loi gaussienne centrée réduite $Y \sim \mathcal{N}(0, I_d)$ et, ensuite, construire le vecteur gaussien $X$ à travers de $X = \Sigma^{1/2}Y$, où $\Sigma^{1/2}$ est une racine carrée inversible de la matrice de covariance $\Sigma$, c'est-à-dire, $\Sigma = \Sigma^{1/2}(\Sigma^{1/2})^T$. Ainsi, cette méthode se compose de deux parties principales: la simulation de variables aléatoires de loi gaussienne et le calcul de la racine carré de $\Sigma$:

#### Simulation de la loi gaussienne centrée réduite par inversion de la fonction de répartition:

Considérons de manière générale qu'on souhaite simuler une loi $X$ dont $F(x) = \mathbb{P}(X \leq x)$ est la fonction de répartition. En considérant l'exercice $T8$ , on définit l'inverse généralisée $F^{-1}$ de $F$ par

$$
F^{-1}(y) = \inf(x\in \mathbb{R} , F(x) \geq y) \quad \hbox{ pour } y \in ]0,1[
$$

Soit $U$ une variable aléatoire de loi uniforme sur $]0,1[$. Alors $F^{-1}(U)$ a même loi que $X$, comme nous l'avons démontré précédemment. Ainsi, pour simuler la loi de $X$, il suffit de générer une variable de loi $\mathcal{U}([0, 1])$ et de connaître l'inverse généralisée de $F(x)$. Dans notre cas, nous pouvons utiliser la fonction  *ppf*  de la bibliothèque [scipy](https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.norm.html) pour calculer l'inverse généralisée de la loi gaussienne.

#### Obtention de la racine carrée de la matrice de covariance:
Nous pouvons obtenir $\Sigma^{1/2}$ en utilisant la décomposition en valeurs propres. Comme $\Sigma$ est une matrice symétrique semi-définie positive, il existe une matrice orthogonale $Q$ et une matrice diagonale $S$, composée des valeurs propres non négatives de $\Sigma$, telles que
$$
\Sigma = Q S Q^T.
$$
Ainsi, comme la racine carrée d’une matrice diagonale $S$ s’obtient simplement en prenant la racine carrée de chacun de ses éléments diagonaux, ce qui donne $\sqrt{S}$, on conclut que $\Sigma = QSQ^T = Q\sqrt{S}(Q\sqrt{S})^T$. Ainsi, la racine carrée de $\Sigma$ s’écrit
$$
\Sigma^{1/2} = Q \sqrt{S},
$$
ce qui satisfait bien la relation $\Sigma^{1/2} (\Sigma^{1/2})^T = \Sigma$.

```python
import numpy as np
from scipy.stats import norm
import matplotlib.pyplot as plt

# Paramètres
d = 3  # dimension
rho = 0.9999  # corrélation entre les variables
N = 10000  # taille de l'échantillon

# ------------------------------
# Simulation de variable gaussienne centrée réduite
u = np.random.uniform(0, 1, (N, d))
z = norm.ppf(u)

# ------------------------------
# Matrice de covariance
sigma = rho * np.ones((d, d)) + (1 - rho) * np.eye(d)

# Racine carrée de la matrice de covariance
eigvals, eigvecs = np.linalg.eigh(sigma)
sqrt_Lambda = np.diag(np.sqrt(eigvals))
sigma_sqrt = eigvecs @ sqrt_Lambda

X = z @ sigma_sqrt.T

# ------------------------------
# Application de la fonction de répartition gaussienne pour obtenir la copule:
U = norm.cdf(X)

# ------------------------------
# Visualisation en 2D des projections
plt.figure(figsize=(10, 4))
for i in range(1, d):
    plt.subplot(1, d - 1, i)
    plt.scatter(U[:, 0], U[:, i], alpha=0.5, s=5)
    plt.title(f"U0 vs U{i}")
    plt.xlabel("U0")
    plt.ylabel(f"U{i}")
    plt.grid()

plt.suptitle("Projections de la copule gaussienne simulée", fontsize=14)

plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.show()
# ------------------------------
```

En variant la correlation, on vérifie, empiriquement, que $(\Phi(X_1),\dots,\Phi(X_d))$ est un vecteur dont la fonction de répartition correspond à la copule gaussienne de matrice de covariance $\Sigma$. Par exemple, pour la corrélation $\rho$ proche de 1, les graphiques de $U_0 \times U_1$ et $U_0 \times U_2$ approximent la diagonale, conforme ce qu'on attendait d'une copule de comonoticité.

#### __Simulation S2:__

*__En dimension $d = 2$, afficher un nuage de 1000 points dans le plan $[0, 1]^2$ simulés selon la copule gaussienne de corrélation $\rho \in [−1, 1]$ $(\Sigma_{11} = \Sigma_{22} = 1, \Sigma_{12} = \Sigma_{21} = \rho)$, pour__*

*$$\rho \in \{−1, −0.99, −0.95, −0.9, −0.8, \dots , 0.8, 0.9, 0.95, 0.99, 1\}$$* 

*__Commenter. Calculer, de façon empirique (utiliser un estimateur Monte-Carlo en donnant un intervalle de confiance), pour chaque $\rho$, la corrélation entre $U_1$ et $U_2$, c’est à dire $\frac{\mathbb{E}[U_1U_2] −1/4}{1/12}$. Cette corrélation est-elle égale à $\rho$?__*


Dans cette simulation, pour estimer la corrélation entre les variables uniformes de la copule, nous considérons un échantillon $ (U_1^{(i)}, U_2^{(i)}) $ pour $ i = 1, \dots, n $ d'observations simulées et nous calculons:

- le produit empirique : $\widehat{\mu_n} = \frac{1}{n} \sum_{i=1}^{n} U_1^{(i)} U_2^{(i)}$
- la corrélation empirique : $\widehat{\rho_n} = \frac{\widehat{\mu_n} - \frac{1}{4}}{1/12}$, où on connaît les distributions marginales $U_1$, $U_2$ $\sim \mathcal{U}([0,1])$.

Ensuite, pour calculer l'intervalle de confiance, on considére la variance empirique de $\rho_n$:
$$\widehat{\sigma_{\rho_n}}^2 = Var( \frac{\widehat{\mu_n} - \frac{1}{4}}{1/12}) = 144 \widehat{\sigma_{\mu_n}}^2$$
$$\widehat{\sigma_{\mu_n}}^2 = \frac{1}{n - 1} \sum_{i=1}^{n} \left( U_1^{(i)} U_2^{(i)} - \widehat{\mu_n} \right)^2$$ 

Donc, par le théorème de la limite central, $\sqrt{n} \frac{\widehat{\rho_n} - \rho}{\sigma_{\rho_n}}$ converge en loi vers une variable aléatoire de loi normale centrée réduit. Ainsi, d'après le résultat le la section 10.3 du polycopié, nous conclusons que l'intervalle de confiance asymptotique au niveau $1-\alpha$ est donnée par la formule ci-dessous, où $\Phi^{-1}$ est la fonction quantile de la loi gaussienne:
   $$
   \left[ \widehat{\rho_n} - \Phi^{-1}(1-\frac{\alpha}{2}) \cdot \frac{12\widehat{\sigma_{\mu_n}}}{\sqrt{n}},\ 
          \widehat{\rho_n} + \Phi^{-1}(1-\frac{\alpha}{2}) \cdot \frac{12\widehat{\sigma_{\mu_n}}}{\sqrt{n}} \right]
   $$


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
import numpy.typing as npt

# ------------------------------
# Simulation d'une copule gaussienne de dimension 2
def simulate_gaussian_copula(rho: npt.NDArray, n: int = 1000) -> npt.NDArray:
    Sigma = np.array([[1, rho], [rho, 1]])

    # Décomposition spectrale pour racine
    eigvals, eigvecs = np.linalg.eigh(Sigma)
    sqrt_Lambda = np.diag(np.sqrt(eigvals))
    sigma_sqrt = eigvecs @ sqrt_Lambda

    # Simulation de la loi gaussienne
    u = np.random.uniform(0, 1, (n, 2))
    z = norm.ppf(u)

    # Simulation de X ~ N(0, sigma)
    X = z @ sigma_sqrt.T

    # Application de la fonction de répartition gaussienne pour obtenir la copule
    U = norm.cdf(X)
    return U


# ------------------------------
# Estimateur de corrélation empirique + IC Monte Carlo
def empirical_correlation_with_ci(U: npt.NDArray, alpha: float=0.05) -> tuple:
    U1, U2 = U[:, 0], U[:, 1]
    prod = U1 * U2
    n = len(prod)
    mu_hat = np.mean(prod)
    var_hat = np.var(prod, ddof=1)
    se = np.sqrt(var_hat / n)

    # Corrélation empirique
    corr_hat = (mu_hat - 0.25) / (1 / 12)

    # Intervalle de confiance à 1 - alpha
    z = norm.ppf(1 - alpha/2)  
    half_width = z * 12* se 
    # IC asymptotique 
    ci_low = corr_hat - half_width
    ci_high = corr_hat + half_width


    return corr_hat, (ci_low, ci_high)


# ------------------------------
# Paramètres
rhos = np.concatenate(
    [
        np.array([-1, -0.99, -0.95, -0.9]),
        np.linspace(-0.8, 0.8, 17),
        np.array([0.9, 0.95, 0.99, 1]),
    ]
)
N = 1000  # nombre d'échantillons pour chaque simulation
alpha = 0.05  # niveau de confiance pour l'IC
results = []

# ------------------------------
plt.figure(figsize=(18, 18))

for i, rho in enumerate(rhos):
    U = simulate_gaussian_copula(rho, N)
    corr_hat, (ci_low, ci_high) = empirical_correlation_with_ci(U, alpha)
    results.append((rho, corr_hat, ci_low, ci_high))

    # Nuage de points
    plt.subplot(5, 5, i + 1)
    plt.scatter(U[:, 0], U[:, 1], s=5, alpha=0.5)
    plt.title(
        f"p = {rho:.2f}\nCorr emp ≈ {corr_hat:.2f}\nIC{100*(1-alpha)}% [{ci_low:.2f}, {ci_high:.2f}]"
    )
    plt.grid()
    plt.xlabel("U1")
    plt.ylabel("U2")
    
    

plt.suptitle(
    "Simulations de copules gaussiennes et corrélations empiriques", fontsize=16
)
plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.show()

# ------------------------------
print("\np théorique | Corr. estimée |  IC 95%")
print("-" * 40)
for rho, corr, lo, hi in results:
    print(f"{rho:10.2f} | {corr:13.4f} | [{lo:.4f}, {hi:.4f}]")
```

#### Commentaires
On observe pour les nuages de points que:
- Pour $ \rho = 1 $, les points sont alignés sur la diagonale : $ U_1 = U_2 $.
- Pour $ \rho = -1 $, les points sont alignés sur la diagonale secondaire : $ U_1 + U_2 = 1 $.
- Pour $ \rho = 0 $, le nuage est homogène dans le carré $ [0,1]^2 $, indiquant une apparente indépendance.
- Pour des valeurs intermédiaires, les points sont concentrés le long d'une diagonale selon le signe et l'intensité de $ \rho $.
  
De plus, on observe que $\widehat{\rho_n}$ s'approxime de $\rho$ à la mesure que la taille de l'échantillon augmente, pourtant, pour $N=1000$, on peut observer encore une discrepance considerable entre la corrélation théorique et la corrélation estimée, pour certains valuers de $\rho$. Par exemple, dans le cas où $\rho = 0$, la corrélation estimée a eu un erreur absolu de l'ordre de $10\%$.

### __Copules archimédiennes__

__On considère $\xi$ une variable aléatoire telle que $P(\xi > 0) = 1$, et on considère pour $t ≥ 0$, $\psi(t) = \mathbb{E}[e^{t\xi}]$ la transformée de Laplace de $\xi$.__

#### __Question T12.__

*__Montrer que $\psi$ est strictement décroissante, continue telle que $\psi(0) = 1$ et $\lim_{t \rightarrow \infty} \psi(t) = 0$. On note $\psi^{−1}:]0, 1] \rightarrow \mathbb{R}^+$ son inverse.__*

Justifions la bonne définition de $\psi$ sur $\mathbb{R}_+$ :

Soit $t \geq 0$,  on peut écrire $\begin{equation} \exp(-t \xi)=\exp(-t \xi)\mathbb{1}_{\xi>0}+\exp(-t \xi)\mathbb{1}_{\xi \leq 0} \end{equation}$.

Or le deuxième terme est nulle sauf sur un ensemble de mesure nulle, donc est d'espérance finie (et même nulle). Le premier terme est lui borné par 1 donc intégrable. $$\fbox{Ainsi, $\psi$ est bien définie.}$$


De plus, soit $t>0$ et $(a_n)_{n\in \mathbb{N}}$ une suite de réels tels que $a_n \underset{n\rightarrow +\infty}{\longrightarrow} t$. La continuité de la fonction exponetielle assure la converge simple de $\exp(-a_n \xi)$ vers $\exp(-t \xi)$.
De plus, ces fonction sont bornées à partir d'un certain rang presque partout par 1. En effet, comme $a_n \underset{n\rightarrow +\infty}{\longrightarrow} t$ et que $t>0$, on peut affirmer qu'à partir d'un certain rang les termes de la suite sont strictemenet positifs. De plus sur un ensemble de probabilité 1, $\xi>0$. Ainsi, on peut appliquer le théorème de Convergence Dominée et affirmer que $\psi (a_n) \underset{n\rightarrow +\infty}{\longrightarrow} \psi(t)$. Cela assure la continuité en $t$. Ceci étant vrai pour tout $t>0$ on peut affirmer que $$\fbox{$\psi$ est continue sur $\mathbb{R}_+^*$.}$$

La stricte décroissance est obtenue à partir du lemme suivant.

**Lemme.**  
Soit $X$ une variable aléatoire réelle et intégrable telle que $\mathbb{P}(X>0)=1$ alors, $\mathbb{E}(X)>0$.

En effet, la décomposition du début de la question nous permet d'affirmer que $\mathbb{E}(X) \geq 0$.
Supposons alors par l'absurde que cette espérance est nulle.
On a : $\begin{equation} 0=\int_{\Omega} X(\omega) dLeb(\omega) \end{equation}$ Or, cette intégrale se calcue sur l'ensemble de mesure totale $\Omega_1$ sur lequel on a $X>0$. Mais alors par les propriétes de l'intégrale de Lebesgue cela signifie que $X=0$ presque surement (sur $\Omega_1$). Ceci est absurde car sur $\Omega_1$ on a $X>0$.


Appliquons ceci pour montrer la stricte décroissance, notons $\Omega_1=\{\omega ; \xi(\omega)>0\}$

Soit $0\leq x<y$ :
$\begin{equation} \forall \omega \in \Omega_1, \space \exp(-x \xi(\omega))-\exp(-y\xi(\omega))>0 \end{equation}$
Donc en utilisant le lemme (car $\mathbb{P}(\Omega_1)=1$), et la linéarité de l'espérance, $\fbox{on obtient bien la stricte décroissance.}$

On a ensuite la convergence presque sure (au moins sur $\Omega_1$) de $t\mapsto \exp(-t\xi)$ vers 0. De plus, presque partout, on a pour tout $t \geq 0$, $|{\exp(-t\xi)}|\leq 1$. On peut donc appliquer le théorème de convergence dominée et alors, 
$$\boxed{\psi(t)\underset{n\rightarrow +\infty}{\longrightarrow} 0}$$

On a enfin clairement $\psi(0)=1$.

#### __Simulation S3__

*__Calculer $\psi$ lorsque $\xi$ suit une loi Gamma $\Gamma(\frac{1}{\theta}, 1)$, avec $\theta > 0$. En dimension $d = 2$, afficher 1000 points dans le plan $[0, 1]^2$ simulés selon cette copule pour $\theta \in \{10^{-2}, 10^{-1}, 1, 10, 100\}$. Commenter. Calculer ensuite comme à la question S2 la corrélation empirique entre $U_1$ et $U_2$ pour ces valeurs de $\theta$. Comparer qualitativement, pour chacun des cinq nuages de points obtenus, avec le nuage de points obtenus à la question S2 qui donne la corrélation entre $U_1$ et $U_2$ la plus proche.__*


Pour la simulation S3, calculons la transformée de Laplace $\psi(t) = \mathbb{E}[e^{-t\xi}]$ d'une loi $\Gamma(\frac{1}{\theta}, 1)$ et son inverse:

La densité de probabilité de la loi $\Gamma(\frac{1}{\theta}, 1)$ est donnée par:
$$
f(x) = \frac{x^{\frac{1}{\theta}-1}e^{-x}}{\Gamma(1/\theta)} \quad \text{pour } x > 0
$$

Ainsi, sa transformée de Laplace se calcule comme:

$$
    \psi(t) = \mathbb{E}[e^{-t\xi}] = \int_0^{\infty} e^{-tx} \frac{x^{\frac{1}{\theta}-1}e^{-x}}{\Gamma(1/\theta)} dx = \frac{1}{\Gamma(1/\theta)} \int_0^{\infty} x^{\frac{1}{\theta}-1} e^{-(t+1)x} dx
$$

Posons :
$$
u = (t + 1)x \quad \Rightarrow \quad x = \frac{u}{t+1}, \quad dx = \frac{du}{t+1}
$$

Après ce changement de variable, l'intégrale devient :
$$
\psi(t) = \frac{1}{\Gamma(1/\theta)} \int_0^{\infty} \left(\frac{u}{t+1}\right)^{(1/\theta)-1} e^{-u} \frac{du}{t+1} = \frac{(t+1)^{-(1/\theta)}}{\Gamma(1/\theta)} \int_0^{\infty} u^{(1/\theta)-1} e^{-u} du
$$

On reconnaît la fonction Gamma $\int_0^{\infty} u^{(1/\theta)-1} e^{-u} du = \Gamma(1/\theta)$ et on obtient le résultat final:

$$\fbox{$\psi(t) = (1 + t)^{-1/\theta}$}$$

La fonction inverse est obtenue en résolvant $\psi(\psi^{-1}(u)) = u$ :
$$
(1 + \psi^{-1}(u))^{-1/\theta} = u \quad \Rightarrow \quad \fbox{$\psi^{-1}(u) = u^{-\theta} - 1$}
$$


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import gamma

# ------------------------------
# Fonctions psi et son inverse pour xi ~ Gamma(1/theta, 1)
def psi(t: float, theta: float) -> float:
    return (1 + t) ** (-1 / theta)

def psi_inv(u: float, theta: float) -> float:
    return u ** (-theta) - 1

# ------------------------------
# Simulation de la copule archimédienne
def simulate_archimedean_copula(theta: float, n: int=1000)-> np.ndarray:
    # xi ~ Gamma(1/theta, 1)
    xi = gamma.rvs(a=1/theta, scale=1, size=n)

    # V1, V2 ~ U[0,1]
    V = np.random.uniform(size=(n, 2))

    # U_i = psi(-log(V_i)/xi)
    logV = -np.log(V)
    U = psi(logV / xi[:, None], theta)
    return U

# ------------------------------
# Corrélation empirique sur [0,1]^2
def empirical_correlation(U: np.ndarray) -> float:
    U1, U2 = U[:, 0], U[:, 1]
    corr = np.cov(U1, U2)[0, 1] / np.sqrt(np.var(U1) * np.var(U2))
    return corr

# ------------------------------
# Simulation pour différentes valeurs de theta
thetas = [1e-2, 1e-1, 1, 10, 100]
N = 1000
emp_corrs_arch = []

plt.figure(figsize=(16, 8))

for i, theta in enumerate(thetas):
    U = simulate_archimedean_copula(theta, N)
    corr_emp = empirical_correlation(U)
    emp_corrs_arch.append((theta, corr_emp))

    # Nuage de points
    plt.subplot(2, 3, i + 1)
    plt.scatter(U[:, 0], U[:, 1], s=5, alpha=0.5)
    plt.title(f"θ = {theta:.2e}\nCorr emp ≈ {corr_emp:.2f}")
    plt.xlabel("U1")
    plt.ylabel("U2")
    plt.grid()

plt.suptitle(
    "Simulations de copules archimédiennes et corrélations empiriques", fontsize=16
)
plt.tight_layout()
plt.show()


# ------------------------------
# Affichage texte des résultats
print("\ntheta vs Corrélation empirique :")
for theta, corr in emp_corrs_arch:
    print(f"theta = {theta:>7.2e} -> Corr emp = {corr:.4f}")
```

#### Commentaires
Les résultats montrent que la dépendance entre $U_1$ et $U_2$ est d’autant plus forte que $\theta$ est grand : pour $\theta = 100$, les points sont très concentrés près de la diagonale. À l’inverse, pour $\theta = 0.01$, la dépendance est presque nulle et la copule se rapproche de l’indépendance. Contrairement à la copule gaussienne, ici la dépendance est plus directement contrôlée par $\theta$, et la corrélation n'atteint que des valeurs positives, mais qui peuvent être très proches de 1, ce qui illustre la capacité des copules archimédiennes à modéliser des dépendances positives extrêmes. En comparaison avec les copules gaussiennes, nous observons que la dispersion entre les points simulés augmente à la mesure que les variables uniformes sont proches de 1, tandis que, pour les copules gaussiennes, la dispersion est plus intense au centre du graphique de $U_1 \times U_2$.

### __Une application au risque de crédit__

__On considère $d = 100$ entreprises qui ont un emprunt et on modélise leur instant de défaut par une loi exponentielle de paramètre $λ = 0,02$. On note ainsi $T_i = −\frac{1}{\lambda}log(U_i)$ l’instant de défaut de la $i$-ème entreprise, et on s’intéresse à $N = \sum_{i=1}^{d} 1_{\{Ti≤T\}}$, le nombre de défauts avant la date $T = 5$.__

#### __Simulation S4.__

*__Calculer en utilisant une méthode de Monte-Carlo, pour différentes $n \in \{5, 10, 30\}$, la valeur de $P(N ≥ n)$ dans les trois cas suivants__*

*__(a) $U_1, \dots ,U_d$ suit la copule archimédienne décrite à la question S3 avec $\theta = 0.1$,__*

*__(b) $1 − U_1, \dots , 1 − U_d$ suit la copule archimédienne décrite à la question S3 avec $\theta = 0.1$,__*

*__(c) $U_1, \dots ,U_d$ suit la copule gaussienne avec $\Sigma_{ii} = 1$ et $\Sigma_{ij} = \rho$ pour $i \neq j$, avec $\rho \in \{0.5, 0.9\}$.__* 

*__Commenter.__*


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import gamma
from scipy.stats import norm

# ------------------------------
# Paramètres
d = 100
lambda_ = 0.02
T = 5
n_vals = [5, 10, 30]
n_sim = 100000


# ------------------------------
def psi(t: float, theta: float) -> float:
    return (1 + t) ** (-1 / theta)


# Fonction de simulation des temps de défaut à partir de U
def simulate_defaults(U: npt.NDArray, T: float | int, lambda_: float) -> npt.NDArray:
    T_i = -np.log(U) / lambda_
    n_defaut = np.sum(T_i <= T, axis=1)
    return n_defaut

# ------------------------------
# Estimateur de P(N >= n)
def estimate_p_ge_n(N: npt.NDArray, n: int) -> float:
    return np.mean(N >= n)

# ------------------------------
# Copule archimédienne (theta=0.1)
def simulate_archimedean_d_dim(theta: float, d: int, n: int) -> npt.NDArray:
    # xi ~ Gamma(1/theta, 1)
    xi = gamma.rvs(a=1/theta, scale=1, size=n)

    # V1, V2 ~ U[0,1]
    V = np.random.uniform(size=(n, d))

    # U_i = psi(-log(V_i)/xi)
    logV = -np.log(V)
    U = psi(logV / xi[:, None], theta)
    return U


# ------------------------------
def simulate_gaussian_d_dim(rho: float, d: int, n: int) -> npt.NDArray:
    sigma = rho * np.ones((d, d)) + (1 - rho) * np.eye(d)

    # Décomposition spectrale pour racine
    eigvals, eigvecs = np.linalg.eigh(sigma)
    sqrt_Lambda = np.diag(np.sqrt(eigvals))
    sigma_sqrt = eigvecs @ sqrt_Lambda

    # Simulation de la loi gaussienne
    u = np.random.uniform(0, 1, (n, d))
    z = norm.ppf(u)

    # Simulation de X ~ N(0, sigma)
    X = z @ sigma_sqrt.T

    # Application de la fonction de répartition Φ pour obtenir la copule
    U = norm.cdf(X)
    return U

   
# ------------------------------
# Exécution pour les 3 cas
results = {}

# (a) Archimédienne theta=0.1
U_a = simulate_archimedean_d_dim(theta=0.1, d=d, n=n_sim)
N_a = simulate_defaults(U_a, T=T, lambda_=lambda_)
results["Arch(theta=0.1)"] = [estimate_p_ge_n(N_a, n) for n in n_vals]

# (b) 1 - Archimédienne theta=0.1
U_b = 1 - simulate_archimedean_d_dim(theta=0.1, d=d, n=n_sim)
N_b = simulate_defaults(U_b, T=T, lambda_=lambda_)
results["1 - Arch(theta=0.1)"] = [estimate_p_ge_n(N_b, n) for n in n_vals]

# (c1) Gaussienne rho=0.5
U_c1 = simulate_gaussian_d_dim(rho=0.5, d=d, n=n_sim)
N_c1 = simulate_defaults(U_c1, T=T, lambda_=lambda_)
results["Gauss(p=0.5)"] = [estimate_p_ge_n(N_c1, n) for n in n_vals]

# (c2) Gaussienne rho=0.9
U_c2 = simulate_gaussian_d_dim(rho=0.9, d=d, n=n_sim)
N_c2 = simulate_defaults(U_c2, T=T, lambda_=lambda_)
results["Gauss(p=0.9)"] = [estimate_p_ge_n(N_c2, n) for n in n_vals]

# ------------------------------
print("Probabilités estimées P(N ≥ n):\n")
header = f"{'n':>5} | " + " | ".join([f"{k:^20}" for k in results.keys()])
print(header)
print("-" * len(header))

for i, n in enumerate(n_vals):
    row = f"{n:5d} | " + " | ".join([f"{results[k][i]:^20.4f}" for k in results])
    print(row)

# Afficher le nombre moyen de défauts N lors d'un événément N >= 5
mean_N = {
    "Arch(theta=0.1)": np.mean(N_a[N_a >= 5]),
    "1 - Arch(theta=0.1)": np.mean(N_b [N_b >= 5]),
    "Gauss(p=0.5)": np.mean(N_c1[N_c1 >= 5]),
    "Gauss(p=0.9)": np.mean(N_c2[N_c2 >= 5]),
}
print("\nNombre moyen de défauts, lors d'un événément N >= 5")
for k, v in mean_N.items():
    print(f"{k:>20}: {v:.2f}")
```

#### Commentaires
Les résultats démontrent l'impact crucial de la structure de dépendance sur la distribution des défauts conjoints. Dans le cas (a) avec une copule archimédienne $(\theta=0.1)$, on observe une probabilité élevée de petits clusters de défauts $(\mathbb{P}(N \geq 5) = 0.9043)$ mais une probabilité quasi-nulle d'événements extrêmes $(\mathbb{P}(N \geq 30) \approx 0)$. Cette configuration, proche de l'indépendance, montre que les défauts surviennent généralement de manière isolée plutôt que coordonnée, avec un nombre moyen modéré d'entreprises en défaut (environ 10) lors des événements $N \geq 5$. À l’inverse, dans le cas (b), la dépendance est anti-monotone, ce qui augmente la probabilité d’observer plusieurs défauts simultanés.

Pour les copules gaussiennes, deux régimes distincts apparaissent :

Avec $\rho=0.5$, le modèle montre un comportement intermédiaire $(\mathbb{P}(N \geq 30)=0.0961)$, pourtant la probabilité de risques extrêmes est plus élévées qu'avec la copule archimédienne, ce qui revèle des limitations des copules gaussiennes pour la modélisation de ce type de phénomène.

Avec $\rho=0.9$, on constate un effet de clustering marqué : le nombre moyen de défauts atteint 44.89 lors des événements $N \geq 5$. Le paradoxe apparent d'une $\mathbb{P}(N \geq 5)=0.2068$ plus faible qu'avec $\rho=0.5$ s'explique par une polarisation entre deux régimes extrêmes - soit presque toutes les entreprises survivent, soit elles font défaut simultanément.

En conclusion, cette analyse illustre que, à marges fixées, le choix de la copule est crucial pour évaluer les risques extrêmes, et notamment pour la modélisation du risque dans un portefeuille de crédit.
